```python
def summ(n):
    if n == 1:
        return 1
    res = n + summ (n-1)
    return res
res = summ(5)
print(res)
```

```python
def fact(n):
    if n == 1:
        return 1
    res = n  * fact(n - 1)

    return res
res = fact(5)
print(res)
```

```python
def gcd(a,b):
    if a<=b:
        b = b % a

        if b == 0:
            return a

        else:
            res = gcd(a,b)
            return res
    else:
        a = a % b

        if a == 0:
            return b
        else:

            res = gcd(a,b)
            return res
result = gcd(246,668)
print(result)
```

```python
def gcd(a,b):
  if b == 0:
    return a
  return gcd(b, a%b)
print(gcd(246,668))

```

```python
def summ(n):
    if n == 0:
        return n
    temp = n% 10
    rem = n //10
    result = summ(rem)
    return temp + result
a = summ(56789)
print(a)
```
